## Q-1. Write an SQL query to print the FIRST_NAME from Worker table after removing white spaces from the right side

```python
import mysql.connector as connection
mydb=connection.connect(host="localhost",user="root",database="Ineuron",passwd="mysql",use_pure=True)
cur=mydb.cursor()
cur.execute("SELECT RTRIM(FIRST_NAME) FROM worker as name;")
print(cur.fetchall())
```

## Q-2. Write an SQL query that fetches the unique values of DEPARTMENT from Worker table and prints its length.

```python
mydb=connection.connect(host="localhost",user="root",database="Ineuron",passwd="mysql",use_pure=True)
cur=mydb.cursor()
cur.execute("SELECT Distinct(Department), LENGTH(Department) as Length FROM worker;")
print(cur.fetchall())
```

## Q-3. Write an SQL query to fetch nth max salaries from a table.

```python
mydb=connection.connect(host="localhost",user="root",database="Ineuron",passwd="mysql",use_pure=True)
cur=mydb.cursor()
cur.execute("SELECT Distinct(salary) from worker order by salary DESC LIMIT 4,1")
print(cur.fetchall())
```
