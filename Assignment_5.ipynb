# Q-1. Write an SQL query to print details of workers excluding first names, “Vipul” and “Satish” from Worker table.

```python
import mysql.connector as connection
mydb=connection.connect(host='localhost',user='root',passwd='mysql',database='ineuron',use_pure=True)
cur=mydb.cursor()
cur.execute('select * from worker where First_name!="Vipul" and First_name!="Satish"')
print(cur.fetchall())
```

# Q-2. Write an SQL query to print details of the Workers whose FIRST_NAME ends with ‘h’ and contains six alphabets

```python
cur.execute('select First_name from worker where First_name like "_____h"')
print(cur.fetchall())
```
