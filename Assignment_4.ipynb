# Q-1. Write an SQL query to print the FIRST_NAME from Worker table after replacing ‘a’ with ‘A’.

```python
import mysql.connector as connection
mydb=connection.connect(host='localhost',user='root',passwd='mysql',database='ineuron',use_pure=True)
cur=mydb.cursor()
cur.execute('select replace(First_name,"a","A") from worker')
print(cur.fetchall())
```

# Q-2. Write an SQL query to print all Worker details from the Worker table order by FIRST_NAME Ascending and DEPARTMENT Descending.

```python
import mysql.connector as connection
mydb=connection.connect(host='localhost',user='root',passwd='mysql',database='ineuron',use_pure=True)
cur=mydb.cursor()
cur.execute('select * from worker ORDER by First_name ASC,DEPARTMENT DESC')
print(cur.fetchall())
```

# Write an SQL query to fetch the names of workers who earn the highest salary.

```python
import mysql.connector as connection
mydb=connection.connect(host='localhost',user='root',passwd='mysql',database='ineuron',use_pure=True)
cur=mydb.cursor()
cur.execute('select First_name,Last_name,salary from worker WHERE salary=(select max(salary) from worker)')
print(cur.fetchall())
```
