## Q-19. Write an SQL query to determine the 5th highest salary without using TOP or limit method.

```python
import mysql.connector as connection
mydb=connection.connect(host="localhost",database="Ineuron",user="root",passwd="mysql",use_pure=True)
cur=mydb.cursor()
cur.execute("select * from worker e1 where 5-1=(select count(distinct salary) from worker e2 where e2.salary>e1.salary)")
print(cur.fetchall())
```

## Q-20. Write an SQL query to fetch the list of employees with the same salary.

```python
import mysql.connector as connection
mydb=connection.connect(host='localhost',user='root',database='ineuron',passwd='mysql',use_pure=True)
cur=mydb.cursor()
cur.execute('Select * from worker where salary in(select salary from worker group by salary having count(*)>1)order by salary asc')
print(cur.fetchall())            
```
