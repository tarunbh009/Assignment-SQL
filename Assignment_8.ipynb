## Q-15. Write an SQL query to fetch intersecting records of two tables.

```python
import mysql.connector as connection
mydb=connection.connect(host='localhost',user='root',passwd='mysql',database='ineuron',use_pure=True)
cur=mydb.cursor()
cur.execute("Select * from worker inner join worker2 on worker.ID=worker2.ID")
print(cur.fetchall())
```

## Q-16. Write an SQL query to show records from one table that another table does not have.

```python
cur.execute("Select * from worker left join worker2 on worker.ID=worker2.ID")
print(cur.fetchall())
```
