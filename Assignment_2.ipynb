## Q-1. Write an SQL query to print the first three characters of FIRST_NAME from Worker table.

```python
import mysql.connector as connection
mydb=connection.connect(host="localhost",user="root",database="Ineuron",passwd="mysql",use_pure=True)
cur=mydb.cursor()
cur.execute("select SUBSTRING(FIRST_NAME,1,3) from worker")
print(cur.fetchall())
```

## Q-2. Write an SQL query to find the position of the alphabet (a) in the first name column Amitabh from Worker table.

```python
cur=mydb.cursor()
cur.execute("SELECT POSITION('A' IN First_name) FROM Worker WHERE First_name='Amitabh'")
print(cur.fetchall())
```

## Q-3. Write an SQL query to print the name of employees having the highest salary in each department.

```python
cur=mydb.cursor()
cur.execute("SELECT Department, max(salary) FROM Worker GROUP by Department")
print(cur.fetchall())
```
