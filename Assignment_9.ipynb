## Q-17. Write an SQL query to show the top n (say 10) records of a table.

```python
import mysql.connector as connection
mydb=connection.connect(host="localhost",database="Ineuron",user="root",passwd="mysql",use_pure=True)
cur=mydb.cursor()
cur.execute("SELECT * from worker limit 10")
print(cur.fetchall())
```

## Q-18. Write an SQL query to determine the nth (say n=5) highest salary from a table.

```python
import mysql.connector as connection
mydb=connection.connect(host="localhost",database="Ineuron",user="root",passwd="mysql",use_pure=True)
cur=mydb.cursor()
cur.execute("select * from worker e1 where 5-1=(select count(distinct salary) from worker e2 where e2.salary>e1.salary)")
print(cur.fetchall())
```
